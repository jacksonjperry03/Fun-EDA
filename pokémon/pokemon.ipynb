## Question 1
What is the most popular primary type for pokemon?

```python
pokemon_df['Primary Type'].value_counts()
```

```python
pokemon_df['Primary Type'].value_counts().plot(kind='bar')
```

Conclusion: Water appears to be the most frequent primary type of pokemon.

## Question 2
What type combinations are the most frequent?

```python
pokemon_df.groupby('Primary Type')['Secondary Type'].value_counts()
```

```python
pd.crosstab(pokemon_df['Primary Type'], pokemon_df['Secondary Type']) #source - https://pbpython.com/pandas-crosstab.html
```

Conclusion: The Normal/Flying type combination is the most frequent.

## Question 3
Does a pokemon's primary type influence speed and health points?

```python
pokemon_df.groupby('Primary Type')[['Speed', 'HP']].describe()
```

```python
sns.scatterplot(data=pokemon_df, x='HP', y='Speed',hue='Primary Type')
```

Conclusion: There does not seem to be a clear relationship between type and health/speed. A pokemon's health points or speed cannot be used to reliably predict it's primary type.
